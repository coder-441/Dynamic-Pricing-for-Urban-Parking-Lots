```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from math import radians, cos, sin, asin, sqrt
import pathway as pw
from bokeh.plotting import figure, show, output_notebook
from bokeh.models import ColumnDataSource
from bokeh.layouts import column
from bokeh.io import push_notebook
import time
```

```python
df = pd.read_csv("/content/dataset.csv")
df['timestamp'] = pd.to_datetime(df['LastUpdatedDate'] + ' ' + df['LastUpdatedTime'], dayfirst=True)
df['VehicleType'] = df['VehicleType'].map({'car': 1, 'bike': 0.5, 'truck': 1.5})
df['TrafficConditionNearby'] = df['TrafficConditionNearby'].map({'low': 1, 'medium': 2, 'high': 3})
df.fillna(0, inplace=True)
```

```python
df['base_price'] = 10.0
df['price_model_1'] = df['base_price']
df['price_model_2'] = df['base_price']
df['price_model_3'] = df['base_price']
```

```python
sample_lot_id = df['SystemCodeNumber'].iloc[0]
lot_data = df[df['SystemCodeNumber'] == sample_lot_id].sort_values('timestamp')

source = ColumnDataSource(data={
    'x': lot_data['timestamp'],
    'y1': lot_data['price_model_1'],
    'y2': lot_data['price_model_2'],
    'y3': lot_data['price_model_3']
})
```

```python
p = figure(title=f"Real-Time Pricing Visualization - {sample_lot_id}", x_axis_type='datetime', width=900, height=400)
p.line('x', 'y1', source=source, color='blue', legend_label='Model 1')
p.line('x', 'y2', source=source, color='green', legend_label='Model 2')
p.line('x', 'y3', source=source, color='red', legend_label='Model 3')
p.legend.location = "top_left"
p.xaxis.axis_label = 'Timestamp'
p.yaxis.axis_label = 'Price ($)'

handle = show(p, notebook_handle=True)
```
